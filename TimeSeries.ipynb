```
import pandas as pd
from scipy.stats import kurtosis
import numpy as np
import seaborn as sns
from IPython.display import display 

cm = sns.light_palette("green", as_cmap=True)

def _calculate_log_normal_return(prices):
  df_temp=np.log(prices/prices.shift(1)).dropna()*100
  return df_temp

def _calculate_Mean(prices):
  df_temp=np.mean(prices.values)
  return df_temp

def _calculate_Std(prices):
  df_temp=np.std(prices.values)
  return df_temp

def _calculate_stats(prices):  
  _stats_dict['Mean']=_calculate_Mean(prices)
  _stats_dict['Standard Deviation']=_calculate_Std(prices)
  print(_stats_dict)
  return (_stats_dict)
# Function that specifies CSS for each cell
def color_fill_conditional(val):
    # Define the styling for a cell based on the value(val)
    styling = 'background-color: yellow;' if val >0 else 'background-color: black; color:yellow;'
    # Return the CSS attribute-value pair
    return styling 


df=pd.read_excel("/content/sample_data/HistoricalData.xlsx")

keys = df['Instrument'].unique()
# keys={'Price','LogNormal'}
df_temp=pd.DataFrame()
_hist_dict={}
_data_dict={}
_stats_prices_dict={}

for key in keys:
  prices=pd.Series((df[df['Instrument']==key]['Price']).values)
  _data_dict[key]=prices
  _hist_dict[key]=_calculate_log_normal_return(prices)#_data_dict

df_new=(pd.DataFrame(_hist_dict)) 
df_prices=(pd.DataFrame(_data_dict)) 
df_prices['Price Date']=df['Price Date'].unique()
df_prices.set_index(df['Price Date'].unique())
_stats_prices_dict['Mean']=pd.Series(pd.DataFrame(_data_dict).mean())
_stats_prices_dict['Std Dev']=pd.Series(pd.DataFrame(_data_dict).std())
_stats_prices_dict['Skewness']=pd.Series(pd.DataFrame(_data_dict).skew())
_stats_prices_dict['Kurtosis']=pd.Series(pd.DataFrame(_data_dict).kurtosis())
corr=df_new.corr()
# corr.style.applymap(color_fill) # Apply the css to each cell
covv=df_new.cov()
# df_new.to_csv("/content/sample_data/Data.csv")
display("*************PRICES******************")
display(df_prices)
print("*************LOG RETURNS******************")
display(df_new)
print("*************STATISTICS******************")
display(pd.DataFrame.from_dict(_stats_prices_dict).T)
print("*************CORRELATION******************")
corr.style.background_gradient(cmap=cm)
display(corr)
print("*************COVARIANCE******************")
# covv.style.applymap(color_fill)
display(covv)
```
