```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# stocks=['AAPL','IBM','MSFT']
stocks=['MSFT','VOW3','ADS.DE','AAPL','IBM']
df_three_Stocks = yf.download(tickers = stocks,start = "2016-04-01",end = "2021-03-31",)
```

```python
df_three_Stocks.head()
```

```python
df_three_Stocks.columns
```

```python
df=pd.DataFrame(columns=['Adj Close',	'Close',	'High',	'Low',	'Open','Volume','Ticker','Date'])
for stock in stocks:
  df_stock=yf.download(tickers = stock,start = "2016-04-01",end = "2021-03-31",)
  df_stock['Ticker']=stock
  df_stock['IDENTIFIER_TYP']='INHOUSE'
  df_stock['Date']=df_stock.index
  df_stock['INSTRUMENT_TYPE']='EQSHR'
  df_stock['PRICE_CURRENCY']='USD'
  df=df.append(df_stock)
df.reset_index(drop=True,inplace=True)
df_new=df[['Ticker','Date','Adj Close',	'Close',	'High',	'Low',	'Open','Volume']]
df_new.to_csv("/content/sample_data/equities.csv")
df
# (IDENTIFIER_TYPE,INSTRUMENT_IDENTIFIER,INSTRUMENT_TYPE,PRICE_CURRENCY,PRICE_DATE_YYYYMMDD,UNIT_PRICE,PRICE_TYPE,MARKET_IDENTIFIER_TYPE,MARKET_IDENTIFIER,ADDITIONAL_PRICE_QUALIFIER,DATA_SOURCE_ID,PRICE_SOURCE,DOWNSTREAM_SYSTEM,PRICING_METHOD,PRICE_QUOTE_METHOD,COMMENTS,Open,Ask,Bid,Close,High,Low,Mid,Last
# INHOUSE,Apple Inc,EQSHR,USD,20191129,66.81,CLOSE,MIC,XNAS,,,,SS_DWDF,,,EQSHR from Investing,66.65,67.0,66.47,66.81,67.0,66.47,66.735,66.81

# )
```
